```python
import pandas as pd
import matplotlib.pyplot as plt

with open('output_0.5.txt', 'r') as file:
    lines = file.readlines()


data = [int(line.strip()) for line in lines[0].split(',')]


# # 创建折线图
plt.plot(data)

# # 添加标题和标签
plt.title('Radius Search (140ms)')
plt.xlabel('point index')
plt.ylabel('included points')

# # 显示图表
plt.show()
```
